```python
# 1) Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva
# una serie con la nota mínima, la máxima, media y la desviación típica.
import ast
#El módulo ast permite convertir un string en un diccionario de forma segura.

def calificaiones():
  notas=input("Ingrese diccionario con las notas de los alumnos de un curso: ")
  cal=[]
  #isinstance permite verificar si notas es un tipo diccionario
  try:
    if not isinstance(notas, dict):
      #se reasigna notas ppara que sea un diccionario, porque antes era un string
      notas=ast.literal_eval(notas)
  # Verificar que todas las claves sean strings y los valores sean números
      for clave, valor in notas.items():
        #clave debe ser un string porque es un nombre
          if valor<0 or valor>5:
              raise ValueError(f"La nota debe estar entre 0 y 5")
          if not isinstance(clave, str):
              raise ValueError(f"Intruduce nombre de estudiante")
          if not isinstance(valor, (int, float)):
              raise ValueError(f"La nota debe ser un número (int o float).")
          cal.append(valor)

      print(f"Nota mínima: {min(cal)}") #mínima nota
      print(f"Nota máxima: {max(cal)}") #máxima nota
      print(f"Nota media: {np.mean(cal)}") #promedio de las notas
      print(f"Desviación estandar: {np.std(cal)}") #desviacion estandar
  except Exception as e:
    print(f"Error: {e}")

calificaiones()
```
